# 01 Download Match History

Fetch historical Leaguepedia Cargo data for seed teams and recursively include opponents.

```python
%load_ext autoreload
%autoreload 2

from moirai.config import CrawlConfig
from moirai.pipeline.download import crawl_match_history

config = CrawlConfig(
    seed_teams=("Top Esports", "Team WE"),
    max_depth=2,
    start_date="2024-01-01",
    limit_per_team=500,
    request_delay_seconds=1,
)

matches = crawl_match_history(config, use_cache=True, verbose=True)
matches.head()
```

```python
from moirai.sources.leaguepedia import LeaguepediaClient

with LeaguepediaClient() as client:
    for seed_team in config.seed_teams:
        candidates = client.find_team_names(seed_team, start_date=config.start_date)
        print(f"Seed: {seed_team}")
        display(candidates.head(10))
```
