# 04 Upcoming Matches

Fetch LoL Esports schedule data and inspect upcoming matches by league.

```python
from moirai.sources.lolesports import LolesportsClient, normalize_leagues

with LolesportsClient() as client:
    leagues = normalize_leagues(client.get_leagues())

leagues.head(30)
```

```python
# Set LEAGUE_ID from the table above, or leave as None for the default schedule feed.
LEAGUE_ID = None

with LolesportsClient() as client:
    upcoming = client.upcoming_events(league_id=LEAGUE_ID)

upcoming.sort_values("start_time").head(50)
```
