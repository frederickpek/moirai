# 02 Build Elo Ratings

Load processed match history, fit team Elo ratings, and inspect current leaders.

```python
import pandas as pd

from moirai.models.elo import fit_elo
from moirai.pipeline.store import load_matches

matches = load_matches()
elo = fit_elo(matches)

ratings = (
    pd.Series(elo.ratings, name="rating")
    .sort_values(ascending=False)
    .rename_axis("team")
    .reset_index()
)
ratings.head(25)
```
